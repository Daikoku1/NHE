```python
def recommand(a, b, c, data = data):    
    # 각 feature별 리스트 생성
    feel = []
    look = []
    taste = []
    smell = []
    
    # 각 feature의 값들은 순서대로 4, 3, 2의 가중치(임의설정)
    for i in [a, b, c]:
        f = list(data[data['name'] == i]['feel'].values[0].keys())
        l = list(data[data['name'] == i]['look'].values[0].keys())
        t = list(data[data['name'] == i]['taste'].values[0].keys())
        s = list(data[data['name'] == i]['smell'].values[0].keys())
        # 각 리스트에 추가 
        feel.extend([f[0]] * 4 + [f[1]] * 3 + [f[2]] * 2)
        look.extend([l[0]] * 4 + [l[1]] * 3 + [l[2]] * 2)
        taste.extend([t[0]] * 4 + [t[1]] * 3 + [t[2]] * 2)
        smell.extend([s[0]] * 4 + [s[1]] * 3 + [s[2]] * 2)
    
    # 모든 맥주의 유사도 값 측정
    # smell & taste 는 2배 가중치(임의설정)
    
    result_dict = {}
    for j in range(len(data)):
        value = 0
        f = list(data.loc[j, 'feel'].keys())
        l = list(data.loc[j, 'look'].keys())
        t = list(data.loc[j, 'taste'].keys())
        s = list(data.loc[j, 'smell'].keys())
        name = data.loc[j, 'name']
        for k in range(3):
            if f[k] in feel:
                value += feel.count(f[k]) * (4 - k)
            if l[k] in look:
                value += look.count(l[k]) * (4 - k)
            if t[k] in taste:
                value += taste.count(t[k]) * (4 - k) * 2
            if s[k] in smell:
                value += smell.count(s[k]) * (4 - k) * 2
        
        result_dict[name] = value
        
        # max_값 찾기(5개 출력)
    a_list =  sorted(result_dict.items(), key=operator.itemgetter(1), reverse=True)
    a_list = a_list[:5]
    
    result = []
    cnt = 0
    for i in a_list:
        cnt += 1
        if i[0] not in [a, b, c]:
            result.append([cnt, i[0]])
    return result
```
